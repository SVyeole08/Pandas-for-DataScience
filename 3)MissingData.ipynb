```python
import numpy as np 
import pandas as pd 
```

### Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()  # Returns number of nan values in a column
```

```python
df.isna().any()  # Returns true which column has nan value
```

### Removing Missing Data

```python
df
```

```python
df.dropna()  # Removes row which has nan value
```

```python
df
```

```python
df.dropna(thresh=3)  # Removes row which has more than 3 nan values
```

### Filling the missing Data

```python
df
```

```python
df.fillna(0)  # Fills the missing data with 0
```

For filling different value in columns:

```python
values = {"A":10,"B":20,"C":30,"D":40}
df.fillna(value=values)
```

**Key takeaway** : When we do **dropna** selection is row wise while we do **fillna** selection is column wise
