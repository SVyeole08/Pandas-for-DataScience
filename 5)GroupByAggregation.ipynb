```python
import numpy as np 
import pandas as pd 
```

**GroupBy**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby("Category")['Sales']  #It makes group of different Categories
for i,j in cat:
    print(i)
    print(j)
```

```python
df.groupby("Category")['Sales'].sum()  #It makes group of different Categories and prints sum of its sales
```

```python
df.groupby("Store")['Sales'].sum()  #It makes of group of different Stores and prints sum of its sales
```

Group by multiple columns

```python
df.groupby(["Store","Category"])["Sales"].sum()
```
