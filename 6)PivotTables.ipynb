```python
import numpy as np
import pandas as pd
```

**Pivot Tables**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
pivot1 = pd.pivot_table(df,values= 'Sales',index = 'Region',columns = "Product")  #Way to create Pivot table; Using pivot table we can make our custom DF.
pivot1
```

```python
pivot2 = pd.pivot_table(df,values="Sales",index="Region",columns="Product",aggfunc="median") #Aggregation function is set to mean by default we can change it like this
pivot2
```

```python
pivot3 = pd.pivot_table(df,values="Sales",index="Region",columns="Product",aggfunc="median", fill_value=0) #fill value fills the given value in missing value
pivot3
```

**Cross Tabs**

```python
pd.crosstab(df['Region'],df['Product'])  #It is used for counting purpose
```

**Key Takeaway** : Pivot table is used for aggregation functions while CrossTab is for counting
