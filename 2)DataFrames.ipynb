```python
import numpy as np 
import pandas as pd 
```

### Creating a DataFrame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)  #Way to create DataFrame by dictionary; Keys of Dictionary are set as column heading
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns=columns)  #If we use other DS as list or array we have to manually set columns otherwise by default it got set to 0 ,1 ,2, and so on
```

```python
df2
```

### Selection and Indexing of Columns

```python
df2
```

```python
#For selecting the column we have to give its label in index braces
df2['Name']  
```

```python
#For selecting multiple columns we have to give their label in form of list in index b_jupyterlab_variableinspector_getshapeof
df2[["Name","Age","Salary"]]
```

### Creating a new column

### Removing Columns

```python
df2
```

By default axis of dataframes set to 0 for selecting columns we have to set axis to 1 because axis 0 only reads row sections

```python
df2.drop("City",axis=1)
```

But after we execute dataframe again the dropped column appear again

```python
df2
```

For avoiding this we have to do:

```python
df2.drop("City",axis=1,inplace=True)
```

```python
df2
```

```python
#For dropping multiple columns we have to give them in list form
df2.drop(["Salary","Name"],axis=1)
```

### Selecting Rows

For removing rows we have to do simply change axis to 0 and write labels of rows 

```python
df2.drop(0,axis=0)
```

```python
df2
```

```python
# For selecting row we have to use .loc attribute
df2.loc[0]
```

```python
# And for multiple rows we have to give rows' labels in listabs
df2.loc[[0,2]]
```

If we had other row labels like any word or something then we can use .iloc attribute to know data of row at location of index

```python
df2.iloc[3]  # 3 is location of 3rd index
```

### Selecting Subsets of Rows and Columns

Same as of matrix slicing in NumPy

```python
df2.loc[[0,1]][["Name","Salary"]]
```

### Conditional Selection

```python
df2
```

People whose Age is greater than 30

```python
df2[df2["Age"]>30]  #Same as Conditionals of NumPy array
```

People whose age is grater than 30 and they should have salary greater than 75000

```python
df2[(df2["Age"]>30) & (df2["Salary"]>75000)]
```
