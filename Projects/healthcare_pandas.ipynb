### Question 8 — Feature Engineering with Dates
Convert `Date of Admission` and `Discharge Date` into real datetime columns using `pd.to_datetime()`. Then create a new column called `Length of Stay` that stores the number of days between admission and discharge for each patient.

```python
df.head()
```

```python
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])
df['Discharge Date']=pd.to_datetime(df['Discharge Date'])
```

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
display(df.head())
```

### Question 9 — apply() with a Custom Function
Write a function called `get_age_group(age)` that returns:
- `"Child"` if age is 12 or under
- `"Teen"` if age is between 13 and 19
- `"Adult"` if age is between 20 and 59
- `"Senior"` if age is 60 or above

Then use `.apply()` to create a new column `Age Group` in `df` using this function.

```python
def get_age_group(age):
    if age < 13:
        return 'Child'
    elif age >=13 and age<=19:
        return 'Teen'
    elif age>=20 and age<=59:
        return 'Adult'
    else:
        return 'Senior'

df['Age Group'] = df['Age'].apply(get_age_group)
df.head()
```

### Question 10 — apply() with a Lambda
Using `.apply()` with a **lambda function**, create a new column called `High Bill` that stores `"Yes"` if a patient's `Billing Amount` is above the *average* billing amount of the whole dataset, and `"No"` otherwise.

```python
average_bill = df['Billing Amount'].mean()
df['High Bill'] = df['Billing Amount'].apply(lambda x: "Yes" if x > average_bill else 'No')
df
```

### Question 12 — GroupBy & Aggregation
