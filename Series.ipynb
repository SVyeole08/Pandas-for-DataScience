## Pandas Series
A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

## Creating Series
There are multiple ways to create a Pandas Series:

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
list = [2,3,4]
arr = np.array([3,4,5])
d = {1:33,2:44,3:55}
```

```python
pd.Series(list)  #here labels are generated by default but we can add custom labels also
```

```python
pd.Series(list,index=labels)  #here custom labels are added
```

```python
pd.Series(arr,index=labels)  #Same with array
```

```python
pd.Series(d)  #As dictionary contains key-value pair then labels are automatically set to keys
```
